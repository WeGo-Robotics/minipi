```python
import rospy
from geometry_msgs.msg import Twist
```

```python
class MiniPiSub:
    def __init__(self):
        rospy.init_node('jupyter_lab_sub')

        self.cmd_sub=rospy.Subscriber("/cmd_vel/auto", Twist, self.cmd_cb)
    
    def cmd_cb(self, msg: Twist):
        print(f"linear.x: {msg.linear.x}, angular.z: {msg.angular.z}", flush=True)
```

```python
if __name__ == "__main__":
    MiniPiSub()
    rate = rospy.Rate(0.25) # 10Hz로 루프 속도 설정
    
    print("ROS Subscriber Node started. Waiting for messages on /cmd_vel...")
    
    while not rospy.is_shutdown():
        rospy.spin() 
        
        rate.sleep()
```
