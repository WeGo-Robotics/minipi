```python
import rospy
from geometry_msgs.msg import Twist
```

```python
class MiniPiPub:
    def __init__(self):
        rospy.init_node('jupyter_lab_pub')
        
        self.cmd_pub=rospy.Publisher("/cmd_vel/auto", Twist, queue_size=10)
        self.cmd_vel=Twist()

    def publish_twist(self):
        self.cmd_vel.angular.z = -0.5
        self.cmd_pub.publish(self.cmd_vel)
```

```python
if __name__ == "__main__":
    MiniPiPub()
    node = MiniPiPub()
    rate = rospy.Rate(10) # 10Hz (0.1초마다 발행)
    
    print("ROS Publisher Node started. Publishing messages to /cmd_vel/auto...")
    
    while not rospy.is_shutdown():
        node.publish_twist()

        rate.sleep()
```
